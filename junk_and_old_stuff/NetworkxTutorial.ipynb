```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
P = nx.petersen_graph()
plt.subplot(121)
nx.draw(P, with_labels=True, font_weight='bold')
plt.subplot(122)
nx.draw_shell(P, nlist=[range(5,10),range(5)])
```

```python
color_map = []
for node in P:
    if node < 5:
        color_map.append('#006874')
    else:
        color_map.append('red')

nx.draw(P, node_color=color_map, with_labels = True)
```
